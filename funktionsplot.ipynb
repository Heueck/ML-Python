Beschriftung

```python
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0,20,0.001)

Ueff=[230,230]
u=325*np.sin(2*np.pi*50*t/1000.0)

plt.title("50 Hz Wechselspannung")
plt.plot(t,u,"b",lw=2, label="$Momentanwert: u(t)$")
plt.plot([0,20],Ueff,"r--", label="$Effektivwert: 230V$")
plt.plot(5,325,"ro",label="$Spitzenwert:325V$")
plt.xlabel("t in ms")
plt.ylabel("u(t) in V")
#plt.legend(loc="upper right")
#plt.legend(loc="lower left")
plt.legend(loc="best")
plt.grid(color="g",ls="dashed", lw="0.5")
plt.show()
```

Mehrere Funktionen in einem Diagramm

```python
import numpy as np
import matplotlib.pyplot as plt
R1,R2,R3=2,4,8
I=0.2
P=1
I=np.linspace(0.1,1,100)

U1=R1*I
U2=R2*I
U3=R3*I
U=P/I

plt.plot(I,U1,I,U2,I,U3,lw=2,color="blue")
plt.plot(I,U,lw=2,color="green")
plt.xlabel("Stromstaerke in A")
plt.ylabel("Spannung in V")
plt.title("Leistungshyperbel U=P/I")
plt.annotate("$R_1$", xy=(1,2),xytext=(+2,-3),textcoords="offset points")
plt.annotate("$R_2$", xy=(1,4),xytext=(+2,-3),textcoords="offset points")
plt.annotate("$R_3$", xy=(1,8),xytext=(+2,-3),textcoords="offset points")
plt.grid(True)
plt.show()
```

Linienstile

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,6.3,500)

y1=np.sin(x)
y3=np.sin(3*x)/3
y5=np.sin(5*x)/5
y7=np.sin(7*x)/7
y=y1+y3+y5+y7

plt.plot(x,y1,color="b",lw=2,linestyle="-")
plt.plot(x,y3,color="r",lw=2,linestyle="--")
plt.plot(x,y5,color="m",lw=2,linestyle=":")
plt.plot(x,y7,color="g",lw=2,linestyle="-.")
plt.plot(x,y,color="black",lw=3)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

Achsenkreuz erzeugen

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-5,5,100)
y=x**2-4
ax=plt.gca()
ax.spines["top"].set_color("none")
ax.spines["right"].set_color("none")
ax.spines["bottom"].set_position(("data",0))
ax.spines["left"].set_position(("data",0))
ticklab=ax.xaxis.get_ticklabels()[0]
ticklab=ax.yaxis.get_ticklabels()[0]
trans=ticklab.get_transform()
ax.xaxis.set_label_coords(6,-0.5,transform=trans)
ax.yaxis.set_label_coords(-0.25,22,transform=trans)
ax.set_xlabel("x")
ax.set_ylabel("f(x)",rotation=True)
plt.plot(x,y,"r",linewidth=2)
plt.show()


```

Logarithmische Skalenteilung

```python
import numpy as np
import matplotlib.pyplot as plt
omega=np.linspace(0.1,100,1000)
for n in range(1,4):
    A=1./(np.sqrt(1.+omega**(2*n)))
    plt.semilogx(omega,A)
plt.xlabel("Frequenz in Hz")
plt.ylabel("A")
plt.grid(True)
plt.show()

```

Polarkoordinaten

```python
import numpy as np
import matplotlib.pyplot as plt
breite,hoehe=5,5 #inch

def theta_rad(winkel1,winkel2):
    theta=[winkel1,winkel2]
    return np.radians(theta)

plt.figure(figsize=(breite,hoehe))
plt.subplot(projection="polar")

plt.polar(theta_rad(0,90),[1,1],"r",lw=3)
plt.polar(theta_rad(90,180),[1,1],"g",lw=3)
plt.polar(theta_rad(180,270),[1,1],"m",lw=3)
plt.polar(theta_rad(270,0),[1,1],"b",lw=3)
plt.polar(theta_rad(0,45),[0,0.6],"black",lw=3)
plt.grid(True)
plt.show()
```

Flchen einfrben

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,5,100)

y1=(x-3)**2
y2=-(x-2)**2+8

plt.plot(x,y1,x,y2,color="black")
plt.fill_between(x,y1,y2,where=y2>=y1,facecolor="b",alpha=0.2)
plt.xlabel("x")
plt.ylabel("y")
plt.show
```
